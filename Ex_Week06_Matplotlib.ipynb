```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

```python
#Sample Dataset
np.random.seed(2)
data = np.random.normal(loc = 100, scale = 10, size = 200)
data
```

```python
#Simple Box Plot
plt.boxplot(data, labels= ['Data'])

plt.title('Simple Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')


plt.grid()
plt.show()
```

```python
#Notched Box Plot
plt.boxplot(data, labels = ['Data'], notch = True)

plt.title('Notched Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
#Pengaturan Marker untuk outlier
plt.boxplot(data, 
            labels = ['Data'], 
            showfliers = True, #Untuk memunculkan outlier
            flierprops = {'markerfacecolor' : 'r', 'marker' : 'd'}) #untuk mengatur warna dan bentuk marker
            
plt.title('Pengaturan Marker')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
#Horisontal Box Plot
plt.boxplot(data, labels = ['Data'], vert = False) #digunakan untuk merubah posisi output menjadi horisontal

plt.title('Horisontal Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
#Multiple Box Plot
np.random.seed(10)

#Pendek atau panjangnya sebaran pada plotting ditentukan oleh standar deviasinya (isi dari parameter ke-2)

data1 = np.random.normal(100,10,200) 
data2 = np.random.normal(80,30,200)
data3 = np.random.normal(90,20,200)
data4 = np.random.normal(70,25,200)

data = [data1,data2,data3,data4]
labels = ['Data 1', 'Data 2', 'Data 3', 'Data 4']
```

```python
plt.boxplot(data, labels = labels) #digunakan untuk merubah posisi output menjadi vertikal

plt.title('Multiple Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
#Sample Dataset
np.random.seed(2)
data = np.random.normal(loc = 100, scale = 10, size = 200)
data
```

```python
#Simple Violin Plot
plt.violinplot(data)

#garis yang dibentuk di sini ini mengindikasikan nilai terbesar dan nilai terkecil dari data sheet yang kita miliki 

plt.title('Simple Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
#Pengaturan pada Violin Plot
plt.violinplot(data,
               showextrema=False, #untuk menampilkan garis tengah yang ada di sini serta garis yang akan mengindikasikan nilai minimum dan nilai maksimum dari data set kita jika bernilai True
               showmeans=False, #untuk memunculkan nilai means berupa garis jika bernilai True
               showmedians=False, #untuk memunculkan nilai medians berupa garis jika bernilai True
               quantiles=None) #untk menampilkan kuartil ke1 hingga ke3 jika parameternya diisi dengan list persentase (misal [0.25, 0.50, 0.75)

plt.title('Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
#Horisontal Violin Plot
plt.violinplot(data, vert=False) #vert digunakan untuk merubah posisi output menjadi horisontal

plt.title('Horisontal Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
#Kasus Twin Axes 1

x=np.linspace(1,10,25) #2 parameter pertama akan menentukan nilai terkecil dan terbesar yang akan dibentuk dari pemanggilan fungsi linspace
                       #Parameter ke3 digunakan untuk menentukan jumlah data Point yang akan dihasilkan dari fungsi linspace
x
```

```python
fig, ax1 = plt.subplots()

ax1.plot(x, np.exp(x), 'bs-', label='exp') #'bs' mengindikasikan warna biru dan marker yang berupa Square dan '-' mengindikasikan adanya garis 
ax1.set_xlabel('Sumbu X')
ax1.set_ylabel('exp')

ax2 = ax1.twinx() #akan menghasilkan akses baru yang akan ditempatkan di atas axes yang pertama
ax2.plot(x, np.log(x), 'ro-', label='log') #'ro' mengindikasikan warna merah dan marker yang berupa bulatan dan '-' mengindikasikan adanya garis 
ax1.set_ylabel('log')

fig.suptitle('Contoh Twin Axes') #Pemanggilan judul
fig.legend(loc='upper left') #posisi dari legendnya

plt.show()
```

```python
#Kasus Twin Axes 2

x=np.random.randint(16,40, size=30) #Parameter pertama akan menentukan nilai terkecilnya
                                    #Parameter kedua akan menentukan nilai terbesarnya
                                    #Parameter ke3 akan menentukan size atau ukuran atau jumlah data Point yang akan dihasilkan oleh fungsi randint
x
```

```python
def C2F(celcius=0):
  return(celcius*1.8) + 32

def konversi_sumbu(ax1):
  y1,y2 = ax1.get_ylim() #y1 akan menampung nilai sumbu y terkecil dari axes yang pertama
                         #y2 akan menampung nilai sumbu y terbesar dari axes yang pertama
  ax2.set_ylim(C2F(y1), C2F(y2))
  ax2.figure.canvas.draw()

fig, ax1 = plt.subplots()
ax2 = ax1.twinx() #akan menghasilkan akses baru yang akan ditempatkan di atas axes yang pertama

ax1.callbacks.connect('ylim_changed', konversi_sumbu)
ax1.plot(x)

ax1.set_xlabel('Hari')
ax1.set_ylabel('Celcius')
ax2.set_ylabel('Fahrenheit')

fig.suptitle('Temperatur Udara') #Pemanggilan judul
plt.show()
```

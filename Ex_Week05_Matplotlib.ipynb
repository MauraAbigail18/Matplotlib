```python
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

```python
#SIMPLE SCATTER PLOT
x=[2,4,6,8,10,11,11.5,11.7] #Menampung nilai integer/float
y=[1,1.5,2,2.5,3,3.5,4,4.5] #Menampung nilai integer/float
```

```python
plt.scatter(x,y, label='Data 1', color ='r') #x dan y akan berkolerasi dengan  posisi tiap data pointsnya

plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh Scatter Plot')

plt.legend()
plt.show()
```

```python
#MULTIPLE SCATTER PLOT
x1=[2,4,6,8,10,11,11.5,11.7] #Menampung nilai integer/float
y1=[1,1.5,2,2.5,3,3.5,4,4.5] #Menampung nilai integer/float

x2=[8,8.5,9,9.5,10,10.5,11] #Menampung nilai integer/float
y2=[3,3.5,3.7,4,4.5,5,5.2] #Menampung nilai integer/float

```

```python
plt.scatter(x1,y1, label='Data 1', color ='r') #x dan y akan berkolerasi dengan  posisi tiap data pointsnya
plt.scatter(x2,y2, label='Data 2', color ='b') #x dan y akan berkolerasi dengan  posisi tiap data pointsnya
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Multiple Scatter Plot')

plt.legend() #akan berkorelasi dengan label yang kita tetapkan untuk setiap scatter plotnya
plt.show()
```

```python
#PENGATURAN MARKER

x1=[2,4,6,8,10,11,11.5,11.7] #Menampung nilai integer/float
y1=[1,1.5,2,2.5,3,3.5,4,4.5] #Menampung nilai integer/float

x2=[8,8.5,9,9.5,10,10.5,11] #Menampung nilai integer/float
y2=[3,3.5,3.7,4,4.5,5,5.2] #Menampung nilai integer/float
```

```python
plt.scatter(x1,y1,
            color = 'cyan', #mengatur warna
            linewidths=1, #tebal garis tepinya
            marker='s', #bentuk marker dari 's' adalah square
            edgecolor='red', #mengatur warna dari garis tepinya
            s=100) #mengatur ukuran dari markernya

plt.scatter(x2,y2,
            color = 'yellow', #mengatur warna
            linewidths=1, #tebal garis tepinya
            marker='^', #bentuk marker dari '^' adalah segitiga
            edgecolor='blue', #mengatur warna dari garis tepinya
            s=200) #mengatur ukuran dari markernya

plt.show()
```

```python
#FUNGSI PLOTTING

def gambar_barplot():
  data1 = [25,85,75,40,60]
  data2 = [40,35,20,55,10]
  kategori = ['A', 'B', 'C', 'D', 'E']

  x = np.arange(len(kategori))
  width = 0.35 #semakin kecil widthnya, semakin tipis juga bar yang akan dihasilkan.
  
  plt.bar(x-width/2,data1,width,label='Data 1')
  plt.bar(x+width/2,data2,width,label='Data 2')

  plt.xticks(x, kategori) #untuk menampilkan isi dari kategori yang kita inginkan dan bukan berisi indeks.
  plt.grid(linestyle='--', #mengatur tipe garis
         linewidth=1, #mengatur tingkat ketebalan garis
         axis='y', #pilihan: x,y,both. Karena disini axis ='y', maka garisnya akan berasosiasi dengan sumbu y
         alpha=0.75) #mengatur tingkat transparansi garis

  plt.xlabel('Kategori')
  plt.ylabel('Jumlah')
  plt.title('Contoh Grouped Bar Plot')

  plt.legend()
  plt.show()
```

```python
gambar_barplot()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

```python
#SIMPLE BAR PLOT
data=[25,45,55,125,225]
kategori=['A','B','C','D','E']
```

```python
plt.bar(kategori,data)
plt.grid()

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

```python
#PENGATURAN GRID DAN COLOR
data=[25,45,55,125,225]
kategori=['A','B','C','D','E']
```

```python
plt.bar(kategori,data, 
        color='red', #mengatur warna
        alpha=0.25) #mengatur tingkat transparansi bar plot
plt.grid(linestyle='--', #mengatur tipe garis
         linewidth=2, #mengatur tingkat ketebalan garis
         axis='y', #pilihan: x,y,both. Karena disini axis ='y', maka garisnya akan berasosiasi dengan sumbu y
         alpha=0.50) #mengatur tingkat transparansi garis

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

```python
#HORISONTAL BAR PLOT
data=[25,45,55,125,225]
kategori=['A','B','C','D','E']
```

```python
plt.barh(kategori,data)

#plt.barh(kategori[::1], data[::1]) -> untuk membalik urutan list

plt.grid(linestyle='--', #mengatur tipe garis
         linewidth= 1, #mengatur tingkat ketebalan garis
         axis='x', #pilihan: x,y,both. Karena disini axis ='x', maka garisnya akan berasosiasi dengan sumbu x
         alpha=0.75) #mengatur tingkat transparansi garis

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Horisontal Bar Plot')

plt.show()
```

```python
#STACKED BAR PLOT
data1 = [25,85,75,40,60]
data2 = [45,35,20,55,10]
kategori=['A','B','C','D','E']
```

```python
plt.bar(kategori,data1,label= 'Data 1')
#plt.bar(kategori,data2,label= 'Data 2') #akan terjadi penimpaan plotting dataset 1 dan 2 pada saat di eksekusi.
plt.bar(kategori,data2,label= 'Data 2',bottom=data1) #parameter bottom menspesifikasikan dataset ke-2 ditempatkan diatas dataset ke-1. 

plt.legend() #berasosiasi dengan parameter label.

plt.grid(linestyle='--', #mengatur tipe garis
         linewidth=2, #mengatur tingkat ketebalan garis
         axis='y', #pilihan: x,y,both. Karena disini axis ='y', maka garisnya akan berasosiasi dengan sumbu y
         alpha=0.75) #mengatur tingkat transparansi garis

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Stacked Bar Plot')

plt.show()
```

```python
#GROUPED BAR PLOT
data1 = [25,85,75,40,60]
data2 = [45,35,20,55,10]
kategori=['A','B','C','D','E']

x = np.arange(len(kategori))
x
```

```python
width = 0.35 #semakin kecil widthnya, semakin tipis juga bar yang akan dihasilkan.
plt.bar(x-width/2,data1,width,label='Data 1')
plt.bar(x+width/2,data2,width,label='Data 2')

plt.xticks(x, kategori) #untuk menampilkan isi dari kategori yang kita inginkan dan bukan berisi indeks.
plt.grid(linestyle='--', #mengatur tipe garis
         linewidth=1, #mengatur tingkat ketebalan garis
         axis='y', #pilihan: x,y,both. Karena disini axis ='y', maka garisnya akan berasosiasi dengan sumbu y
         alpha=0.75) #mengatur tingkat transparansi garis

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Grouped Bar Plot')

plt.legend()
plt.show()
```

```python
#SIMPLE PIE PLOT

kategori= ['A','B','C','D','E']
data = [215,130,245,210,225]
```

```python
plt.pie(data, #berasosiasi dengan sekumpulan nilai integer.
        labels=kategori, #berasosiasi dengan parameter kategori dan sifatnya opsional.
        autopct='%1.1f%%', #digunakan untuk menampilkan presentase dari setiap area pada Pie Plot ini dan sifatnya juga opsional.
        startangle=90) #digunakan untuk melakukan rotasi pada Pie Plot.

plt.title('Contoh Pie Plot') #untuk menampilkan judul pada Pie Plot

plt.show()
```

```python
#PENGATURAN LEGEND DAN COLOR
kategori= ['A','B','C','D','E']
data = [215,130,245,210,225]
warna = ['pink','cyan','#e03364','yellowgreen','skyblue'] #merepresentasikan warna dari setiap isi parameter kategori
```

```python
plt.pie(data, #berasosiasi dengan sekumpulan nilai integer.
        labels=kategori, #berasosiasi dengan parameter kategori dan sifatnya opsional.
        colors=warna, #berasosiasi dengan parameter warna.
        autopct='%1.1f%%', #digunakan untuk menampilkan presentase dari setiap area pada Pie Plot ini dan sifatnya juga opsional.
        startangle=90) #digunakan untuk melakukan rotasi pada Pie Plot.

plt.legend() #menampilkan keterangan warna pada plotting
plt.title('Contoh Pie Plot') #untuk menampilkan judul pada Pie Plot

plt.show()
```

```python
#EXPLODED PIE PLOT

kategori= ['A','B','C','D','E']
data = [215,130,245,210,225]
warna = ['pink','cyan','#e03364','yellowgreen','skyblue'] #merepresentasikan warna dari setiap isi parameter kategori
explode_var =[0.,0.,0.2,0.,0.8] #menampung nilai pecahan yang berkorelasi dengan kategori.
```

```python
plt.pie(data, #berasosiasi dengan sekumpulan nilai integer.
        labels=kategori, #berasosiasi dengan parameter kategori dan sifatnya opsional.
        colors=warna, #berasosiasi dengan parameter warna.
        explode = explode_var, #berkolerasi dengan parameter explode_var.
        autopct='%1.1f%%', #digunakan untuk menampilkan presentase dari setiap area pada Pie Plot ini dan sifatnya juga opsional.
        startangle=90) #digunakan untuk melakukan rotasi pada Pie Plot.

plt.title('Contoh Exploded Pie Plot') #untuk menampilkan judul pada Pie Plot

plt.show()
```

```python
#SAMPLE DATA
data_uas =[
           ['Bejo', 70],
           ['Tejo', 83],
           ['Cecep', 62],
           ['Wati', 74],
           ['Karti', 71]

           ]
```

```python
data_uas
```

```python
#TABLE PLOT DENGAN OO STYLE
fig, ax = plt.subplots()

table = plt.table(cellText=data_uas,
                  loc='center') #berkaitan dengan penempatan tablenya.

table.set_fontsize(14) #merupakan attribute dari object table dan berfungsi untuk mengatur ukuran font.
table.scale(1,4) #ukuran kolom dan baris -> semakin besar semakin besar nilai dilewatkan maka semakin besar pula ukuran yang akan dihasilkan.

ax.axis(False) #digunakan agar axisnya tidak muncul.

plt.show()
```

```python
#TABLE PLOT DENGAN pyplot STYLE

table = plt.table(cellText=data_uas,
                  loc='center') #berkaitan dengan penempatan tablenya.

table.set_fontsize(14) #merupakan attribute dari object table dan berfungsi untuk mengatur ukuran font.
table.scale(1,4) #ukuran kolom dan baris -> semakin besar semakin besar nilai dilewatkan maka semakin besar pula ukuran yang akan dihasilkan.

ax = plt.gca() #digunakan untuk mendapatkan akses yang saat ini sedang aktif dan kependekkan dari get current access.
ax.axis(False) #digunakan agar axisnya tidak muncul.

plt.show()
```
